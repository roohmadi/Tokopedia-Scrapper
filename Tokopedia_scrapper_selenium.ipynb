```python
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.by import By
from selenium import webdriver
import time

from datetime import datetime
from datetime import date

today = date.today()
# Month abbreviation, day and year	
date_str = today.strftime("%d%b%y")
now = datetime.now()
# dd/mm/YY H:M:S
time_str = now.strftime("%H%M")

driver = webdriver.Firefox()

keyword ='mie ashitaki'
market_place = "tokopedia"
print(keyword + "_" + market_place + "_" + date_str + "_" + time_str)
```

```python
#driver.get('https://www.tokopedia.com/search?navsource=&ob=5&srp_component_id=02.01.00.00&srp_page_id=&srp_page_title=&st=product&q=arduino')
driver.get('https://www.tokopedia.com/search?navsource=&ob=5&srp_component_id=02.01.00.00&srp_page_id=&srp_page_title=&st=product&q=' + keyword )
    
counter_page = 0
datas = []
print("Ready")
```

```python
now = datetime.now()

current_time = now.strftime("%H:%M:%S")
print("Current Time Start =", current_time)

while counter_page < 57:
    for _ in range(0, 6500, 500):
        time.sleep(0.5)
        driver.execute_script("window.scrollBy(0,500)")
        
    elements = driver.find_elements(by=By.CLASS_NAME, value='css-12sieg3')
    for element in elements: 
        try:
            img_url = element.find_element(by=By.CLASS_NAME, value='css-1c345mg').get_attribute('src')
        except NoSuchElementException:
            img_url = '0'
        try:
            product_name = element.find_element(by=By.CLASS_NAME, value='css-svipq6').text
        except NoSuchElementException:
            product_name = '0'
        try:
            price = element.find_element(by=By.CLASS_NAME, value='css-1ksb19c').text            
        except NoSuchElementException:
            price = '0'
        try:
            city = element.find_element(by=By.CLASS_NAME, value='css-1kdc32b').text
        except NoSuchElementException:
            city = '0'
        
        try:
            rating = element.find_element(by=By.CLASS_NAME, value='css-t70v7i').text
        except NoSuchElementException:
            rating = '0'
        try:
            url_text = driver.find_element(by=By.CLASS_NAME, value='css-1f2quy8').text

            url = element.find_element(by=By.CLASS_NAME, value='css-1f2quy8').text
        except NoSuchElementException:
            url = '0'
        
        try:
            quantity_sold = element.find_element(by=By.CLASS_NAME, value='css-1duhs3e').text
        except NoSuchElementException:
            quantity_sold = '0'
           
        try:
            shop_name_text = element.find_element(by=By.CLASS_NAME, value='css-1rn0irl').text
            shop_name = (shop_name_text.split('\n'))
            shop_name = shop_name[1]
        except NoSuchElementException:
            shop_name = '0'
        
        #presentase_diskon = element.find_element(by=By.CLASS_NAME, value='css-1qtulwh').text
        #print([price + " - " + (city) + " - " + (shop_name)])
        datas.append({
            'img_url': img_url,
            'product_name': product_name,
            'price': price,
            'city': city,
            'location': city,
            'rating': rating,
            'url': url,
            'quantity_sold': quantity_sold,
            'shop_name': shop_name
            #'presentase_diskon': presentase_diskon
        })
    counter_page += 1
    try:
        next_page = driver.find_element(by=By.XPATH, value="//button[@class='css-1ix4b60-unf-pagination-item' and text()='" + str(counter_page + 1) + "']")
        next_page.click()
    except NoSuchElementException:
        pass
                
print("finish...")
now = datetime.now()
current_time = now.strftime("%H:%M:%S")
print("Current Time Stop=", current_time)
```
