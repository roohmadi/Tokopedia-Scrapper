```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#matplotlib inline
import seaborn as sns

keyword ='mie ashitaki'
market_place = "tokopedia_test"
print(keyword + "_" + market_place)
```

```python
file_name ='mie ashitaki_tokopedia_15Jan23_20351'
df = pd.read_csv(file_name + ".csv") #baca dataset
df
```

```python
#menghitung jumlah produk yang dijual untuk tiap subkategori
df['quantity_sold'].count()
df['shop_name'].count()
#produk_paling_banyak_dijual = df['quantity_sold','shop_name'].count()
#produk_paling_banyak_dijual
```

```python
city_sum_sold = (df[['city', 'quantity_sold']].groupby(['city']).sum()).sort_values(by='quantity_sold',ascending=False).reset_index()
#city_sum_sold = df[['city', 'quantity_sold']].groupby(['city']).sum()
```

```python
sns.barplot(x='city',y='quantity_sold',data=city_sum_sold[:20],color="blue")
plt.xlabel("Penjualan by location")
plt.ylabel("Product Sold")
plt.title("Lokasi Toko Paling Banyak Menjual " + keyword + " di Tokopedia")
plt.xticks(rotation='vertical')
plt.show()
```
